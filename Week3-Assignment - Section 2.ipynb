```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

wiki = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
soup = BeautifulSoup(wiki.content, 'html.parser')

# Get all the tables
tables = soup.find_all('table',class_="wikitable")
values=[]
for table in tables:
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        values.append([ele for ele in cols if ele]) # Get rid of empty values

del values[0] # Get rid of the empty row.
values
```

# Put the list into a dataframe

```python
header_list = ['PostalCode','Borough','Neighborhood']
df = pd.DataFrame(data = values, columns=header_list)
df
```

# Drop all the rows where Borough is not assigned

```python
df.drop(df[df.Borough == 'Not assigned'].index, inplace=True)
df.head()
```

# Replace Neighborhoods that aren't assigned with the name of the Borough

```python
df.Neighborhood = np.where(df.Neighborhood == "Not assigned",df.Borough,df.Neighborhood)
df.head()
```
